```python
genename = "Vwf"
row = plot_df.filter(items = [genename], axis=0)
row
```

```python
dataset = "ACZ"
# select columns by regular expression
data = row.filter(regex=dataset, axis=1)
timepoints = set([name.split("_")[2] for name in data.columns])
celltypes = set([name.split("_")[0] for name in data.columns])
result = {}
"""
result = {
    "E18.5": {
        "Adventitial fibroblast": 0.521484,
        "Alveolar type II": 0.0
        ....
    },
    "P7": {
    "Adventitial fibroblast": 0.1234,
        "Alveolar type II": 0.2345
        ....
    }
}
"""
###
for timepoint in timepoints:
    # get the items in this time point regardless of the cell type
    timepoint_data = data.filter(regex=timepoint, axis=1)
#     print("timepoint is:",timepoint)
    # print("data is: ",timepoint_data)
    result[timepoint] = {}
    for cell in celltypes:
        celltype_data = timepoint_data.filter(regex="^{}_".format(cell), axis=1)
        # if the cell is missing tin thsi time point
        if len(celltype_data.values[0]) == 0:
            result[timepoint][cell] = 0
        else:
            result[timepoint][cell] = celltype_data.values[0][0]
        # print("celltype is:",cell)
        # print("cell type value is:",celltype_data.values[0])
print(result)
```
