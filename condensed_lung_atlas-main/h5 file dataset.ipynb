#### Car4, Vwf, Col1a1, Ptprc, Ms4a1 (genes)

```python
test_df
```

```python
# convert index from binary to string
new_index = []
for i in test_df.index:
    new_index.append(i.decode('utf-8'))
```

```python
# convert column name from binary to string
new_column_name = []
for i in test_df.columns:
    new_column_name.append(i.decode('utf-8'))
```

```python
test_df.index=new_index
test_df.columns=new_column_name
```

```python
test_df.filter(items = ['Car4','Vwf', 'Col1a1', 'Ptprc', 'Ms4a1'], axis=0)
```

```python
# column and index names are in bytes (not string)
# to convert string to byte
# do this
string_val = 'hello'
byte_val = bytes(string_val, 'utf-8')
print(byte_val)

# reverse operation
original_string_val = byte_val.decode('utf-8')
print(original_string_val)
```
